```python
import pandas as pd
```

```python
#FB exports the insights by maximum 180 days. So, I exported 3 files per year, total 6 files for years 2019 & 2020
bc_20191=pd.read_excel(r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\2019-1.xlsx', engine='openpyxl')
bc_20192=pd.read_excel(r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\2019-2.xlsx', engine='openpyxl')
bc_20193=pd.read_excel(r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\2019-3.xlsx', engine='openpyxl')
bc_20201=pd.read_excel(r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\2020-1.xlsx', engine='openpyxl')
bc_20202=pd.read_excel(r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\2020-2.xlsx', engine='openpyxl')
bc_20203=pd.read_excel(r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\2020-3.xlsx', engine='openpyxl')
```

```python
#I merge the 6 dfs for years 2019-2020
bc_2019_2020=pd.concat([bc_20191,bc_20192,bc_20193,bc_20201,bc_20202,bc_20203])
```

```python
bc_2019_2020.head()
```

```python
bc_2019_2020.info()
```

```python
#I create 3 more columns with only the year, month and name of weekday
bc_2019_2020['Year']=bc_2019_2020['Date'].dt.year
bc_2019_2020['Month']=bc_2019_2020['Date'].dt.month_name()
bc_2019_2020['Month_Num']=bc_2019_2020['Date'].dt.month
bc_2019_2020['Weekday']=bc_2019_2020['Date'].dt.day_name()
```

```python
bc_2019_2020.head()
```

```python
bc_2019_2020.info()
```

```python
#I save the df bc_2019_2020 in excel file 
bc_2019_2020.to_excel(r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\Final work\bc_19-20.xlsx')
```

```python
bc_2019_2020_month=bc_2019_2020.groupby(['Year', 'Month_Num', 'Month'])[['Daily Total Reach', 'Daily Organic Reach', 'Daily Paid Reach',
                                                         'Daily Page Engaged Users', 'Daily Total Impressions', 
                                                        'Daily Organic impressions', 'Daily Paid Impressions']].sum()
```

```python
bc_2019_2020_month
```

```python
#I convert the index to columns
bc_2019_2020_month.reset_index(inplace=True)
bc_2019_2020_month
```

```python
#I sort the df by month
bc_2019_2020_month.sort_values(['Year','Month_Num'], inplace=True)
```

```python
bc_2019_2020_month
```

```python
#group by year and day of week
bc_2019_2020_day=bc_2019_2020.groupby(['Year', 'Weekday'])[['Daily Total Reach', 'Daily Organic Reach', 'Daily Paid Reach',
                                                         'Daily Page Engaged Users', 'Daily Total Impressions', 
                                                        'Daily Organic impressions', 'Daily Paid Impressions']].sum()
```

```python
#convert the index to columns
bc_2019_2020_day.reset_index(inplace=True)
bc_2019_2020_day
```

```python
#I save the 3 dataframes bc_2019_2020, bc_2019_2020_month and bc_2019_2020_day in excel file with 3 sheets
path=r'C:\Users\user\Desktop\Anastasios\Προσωπικά έγγραφα\Καρναβάς\FB Bubble Cleaning\Final work\bc_19-20.xlsx'
writer = pd.ExcelWriter(path, engine = 'xlsxwriter')
bc_2019_2020.to_excel(writer, sheet_name = 'All_page', index=False)
bc_2019_2020_month.to_excel(writer, sheet_name = 'per month_page', index=False)
bc_2019_2020_day.to_excel(writer, sheet_name = 'per day_page', index=False)
writer.save()
writer.close()
```
